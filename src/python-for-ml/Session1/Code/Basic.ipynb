### For Formating

```python
x = 20
y = 30
print("this is x value = ",x,"and y value is = ",y)
print("this is x {} and  y {}".format(x,y))
print("this is x {1} and  y {0}".format(x,y))
print("this is x {1} and  y {0}".format('Randa',y)) # we can use any type of argument
print(f"this is x = {20} and  y = {30}") #common or acceptable using (f"") for formating
```

### String Data Type 

```python
name = "Depi R4"
name[-1] # the last char
```

```python
name = "Depi R4"
name[0:4:2] #Slicing with step 2 
```

```python
name = "welcome into depi r4"
upper_txt = name.upper() # Convert all letters to upper case
print(upper_txt)
```

```python
lower_txt = upper_txt.lower() # convert all letters to lower case
print(lower_txt)
```

```python
name = "welcome into depi r4"
name.replace("r4","r5")
```

```python
name = "welcome into depi r4"
x = "r4"
y = "r5"
name.replace(x,y)
```

```python
name = "welcome into depi r4 r4"
name.replace("r4","r5")
```

```python
name = "welcome into depi r4"
name.capitalize() # capitalize the first letter only in the sentence or paragraph
```

```python
name = "welcome into depi r4"
name.title() # capitalize the first letter in all words in the sentence or paragraph
```

```python
name = "welcome into depi r4 r4"
name.count("r4") # count how much this appear in the sentence
```

```python
name = "welcome into depi r4"

''' 
split() 
Method returns list 
We can use it for tokenization in NLP

'''
list_1 = name.split() 
print(type(list_1))
```

```python
sent = " ".join(list_1) # convert it to sentence again
print(sent)
type(sent)
```

```python
sent.startswith("welcome")
```

```python
sent.endswith("depi")
```
