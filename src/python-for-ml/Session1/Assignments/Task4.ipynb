## Encoded Message: ##$$$@!yalpstcejorp EPUVT****9887

### Steps

#### 1. Extract the core part of the message: "yalpstcejorp EPUVT".

```python
message = "##$$$@!yalpstcejorp EPUVT****9887"

core_part = message[7:25]

print("Core part:",core_part)
```

#### 2. Reverse the first word: "yalpstcejorp" becomes "projectsplay".


```python
word1 , word2 = core_part.split()

decoded_word1 = word1[::-1]

print(decoded_word1)
```

#### 3. Replace shifted vowels in the second word: "EPUVT": Replace E->A, U->O to get "APOVT"

```python
decoded_word2 = word2.replace('E','A').replace('U','O')

print(decoded_word2)
```

#### 4. Final decoded message: "projectsplay APTOV".

```python
final_decoded_message = decoded_word1 + " " + decoded_word2

print(final_decoded_message)
```
