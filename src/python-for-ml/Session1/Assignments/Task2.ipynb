## Encoded Message: ###!!@mocleW EPGTQ!!!6789

### Steps

#### 1. Extract the core part of the message: "mocleW EPGTQ".

```python
message = "###!!@mocleW EPGTQ!!!6789"

core_part = message[6:18]

print("Core part:",core_part)
```

#### 2. Reverse the first word: "mocleW" becomes "Welcom".

```python
first_word = core_part[:6]

reversed_word = first_word[::-1]

reversed_word
```

#### 3. Replace shifted vowels in the second word:


```python
'''  
vowel_shift = {'E':'A', 'I':'E', 'O':'I', 'U':'O', 'A':'U'}
So we will replace E to A in EPGTQ

'''
second_word = core_part[6:]

decoded_second_word = second_word.replace('E','A')

print(decoded_second_word)
```

#### 4. Final decoded message: "Welcom APGTQ".

```python
final_decoded_message = reversed_word + " "+ decoded_second_word

print(final_decoded_message)
```
