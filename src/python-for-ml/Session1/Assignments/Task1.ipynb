## Suppose you have this email address “Amit_ml@gmail.edu”

### 1-Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol. If it's not a valid email, return "Invalid email".


```python
email = 'Amit_ml@gmail.edu'

x = email.count("@")

index_after_symbol = (email.index("@") + 1)

email_2 = email[index_after_symbol:] # Get the part after @

y = email_2.count(".")


if (x == 1 and y >=1):
    print("Valid email")
else:
    print("Invalid email")
```

### Extract Username: Extract and return the part of the email before the "@" symbol.


```python
email = 'Amit_ml@gmail.edu'

user_name = email[:7]

print("Username:",user_name)
```

### Extract Domain: Extract and return the domain (the part between "@" and the last ".").


```python
email = 'Amit_ml@gmail.edu'

domain = email[8:13]

print("Domain:",domain)
```

### Check for Domain Ending: Check if the email ends with ".com". If it does , return "Commercial Domain". If it ends with ".edu", return "Educational Domain". Otherwise, return "Other Domain".


```python
email = 'Amit_ml@gmail.edu'

ends = email[-4::1]

if (ends == ".com"):
    print("Commercial Domain")
elif(ends == ".edu"):
    print("Educational Domain")  
else:
    print("Other Domain")  
```
