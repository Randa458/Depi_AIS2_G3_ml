#### If Condition

```python
a = 200
b = 33

if b > a:
    print("b is greater than a")
else:
    print("b is not greater than a")    
```

```python
a = 200
b = 33

if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("b is not greater than a")  
```

```python
x = 5
y = 10

if x > 0 and y > 0:
    print("both x and y are positive numbers.")
```

```python
x = -1
y = 10

if x > 0 and y > 0:
    print("both x and y are positive numbers.")
```

```python
x = -5
y = 10

if x > 0 or y > 0:
    print("both x and y are positive numbers.")
```

```python
x = 5

result = "Even" if x % 2 == 0 else "Odd" # Short hand if
print(result)
```

```python
a = 10
b = 5

result = a + b if a > b else a - b
print(result) 
```

```python
is_sunny = True
temerature = 25

if is_sunny:
    print("It is a sunny day")
    
    if temerature > 20:
        print("The weather is warm")
    else:
        print("The weather is cool")
else:
    print("It is not a sunny day")
```

```python
degree = int(input("Enter your degree between (0:100):"))
 
if 0 < degree <= 100:
    if 90 <= degree <= 100: 
        print('A')
    elif 80 <= degree < 90:
        print("B")
    elif 70 <= degree < 80:
        print("C")
    elif 60 <= degree < 70: 
        print("D")
    else:
        print("F")  
else:
    print("this is in correct") 
```
