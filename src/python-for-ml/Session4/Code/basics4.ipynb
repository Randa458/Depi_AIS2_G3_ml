```python
def fact_num(num : int):
    
    if num == 0 or num ==1:
        return 1
    return num * fact_num(num-1)
    
```

```python
result = fact_num(5)
result
```

```python
def is_prime(num):
    for i in range(2 , num):
        if num % i == 0:
            return False
        return True
```

```python
is_prime(3)
```

```python
def common_factors(num1 : int , num2 : int):
    
    minimum = min(num1 , num2)
    common = []
    
    for i in range(1, minimum + 1):
        if num1 % i == 0 and num2 % i == 0:
            common.append(i)
    return common
```

```python
a = int(input("enter the first number: "))
b = int(input("enter the second number: "))

common_factors(a,b)
```

```python
import random

def generate_password(length: int):
    """
    This function is used to generate a random password of given length.
    Includes uppercase, lowercase, digits, and symbols.
    
    param_1 : the length of password i want to generate
    type param_1 : int
    
    return : password generated
    type return : str
    
    """
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ""
   
    for _ in range(length):
        password += random.choice(chars)  
    return password


```

```python
def count_words(words_list):
    """
    This function is used for calculating number of every unique word in the list
    
    param_1 : list of words i want to calculate from it 
    type param_1 : list
    
    return : return every unique word with count of it in the list
    type return : dict
    
    """
    unique_words = set(words_list)
    count = {word: 0 for word in unique_words}
    for i in unique_words:
        for j in words_list:
            if i == j:
                count[i] += 1
    return count
```

```python
import random

# Responses dictionary
responses = {
    "hello": ["Hello!", "Hi there!", "Greetings!"],
    "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
    "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
    "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
}

# Function to get response
def get_response(user_input):
    for key in responses:
        if key in user_input:
            return random.choice(responses[key])
    return random.choice(responses["default"])

# Chatbot function
def chatbot():
    print("Hello! I'm a simple ChatBot. Type 'goodbye' to quit.")
    
    while True:
        user_input = input("User: ").lower()
        response = get_response(user_input)  
        print("ChatBot:", response)
        
        if user_input == "goodbye":
            break

# Run the chatbot
if __name__ == "__main__":
    chatbot()
```

```python
import random

class PasswordGen:
    """
    This class helps the user generate a random password.
    """
    def __init__(self, length=8):
        self.length = length
        self.char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    
    def gen_pass(self, length=None):
        if length is None:
            length = self.length
        password = ""
        num_of_char = len(self.char)
        for _ in range(length):
            rand_index = random.randrange(0, num_of_char)
            password += self.char[rand_index]
        return password

if __name__ == "__main__":
    gen = PasswordGen()
    rand_password = gen.gen_pass(12)  # generate a password of length 12
    print("Generated Password:", rand_password)
```

```python
class WordCounter:
    """
    This class is used for counting the occurrences of each unique word in a list.
    """
    def __init__(self, words_list):

        self.words_list = words_list
    
    def count_words(self):
        """
        Count the occurrences of each unique word in the list.
        
        return: dictionary with words as keys and counts as values
        type return : dict
        """
        unique_words = set(self.words_list)
        count = {word: 0 for word in unique_words}
        for word in unique_words:
            for w in self.words_list:
                if w == word:
                    count[word] += 1
        return count

if __name__ == "__main__":
    words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
    counter = WordCounter(words)
    counts = counter.count_words()
    print(counts)
```

```python
import random

class SimpleChatBot:
    """
    A simple chatbot that responds to user input based on predefined responses.
    """
    def __init__(self):
        
        self.responses = {
            "hello": ["Hello!", "Hi there!", "Greetings!"],
            "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
            "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
            "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }
    
    def get_response(self, user_input):
        """
        Returns a response based on user input.
        
        """
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        return random.choice(self.responses["default"])
    
    def start_chat(self):
        
        print("Hello! I'm a simple ChatBot. Type 'goodbye' to quit.")
        
        while True:
            user_input = input("User: ").lower()
            response = self.get_response(user_input)
            print("ChatBot:", response)
            
            if user_input == "goodbye":
                break

if __name__ == "__main__":
    bot = SimpleChatBot()
    bot.start_chat()
```
