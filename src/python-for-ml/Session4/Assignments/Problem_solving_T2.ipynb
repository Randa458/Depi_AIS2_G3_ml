### 1. Two Sum

```python
class TwoSum:
    def __init__(self , array : list , target : int ):
        self.array = array
        self.target = target
        
    def two_sum(self):
        for i in range (len(self.array)):
            for j in range (i+1 , len(self.array)):
                if self.array[i] + self. array[j] == self.target:
                    return i , j
            
        return None      

            
    
ts = TwoSum([2, 7, 11, 15], 9)
print(ts.two_sum())
       
            
    
```
