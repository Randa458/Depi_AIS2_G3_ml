## Task 1: Factorial of a Number

```python
class FactorialCalculator:
    """
    This class is used to calculate factorial of a number
    
    Attribute_1 : The number to calculate factorial for.
    type attribute_1 : int
    
    """
    def __init__(self, num : int):
        self.num = num
    def fact_calc(self , num: int):
        """
        This function is used to calculate factorial of a num using recursion
        
        param_1 : The number to calculate factorial for.
        type param_1 : int
        
        return : factorial of a num 
        type return : int
        
        """
        if num == 0 or num == 1:
            return 1
        else:
            return num * self.fact_calc(num-1)
        
obj = FactorialCalculator(5)
result = obj.fact_calc(5)
print(result)
    
```

## Task 2: Prime Number Checker

This task checks whether a given number is a prime number.

```python
class PrimeChecker:
    """
    This class is used to check whether a given number is a prime or not
    
    Attribute_1 : the number i want to check
    type attribute_1 : int
    
    """
    def __init__(self , num : int):
        self.num = num
        
    def is_prime(self):
        """
        This function used to check whether a given number is a prime or not.
        
        return : num is prime or not .
        type return : boolean
        
        """
        for i in range (2 , self.num):
            if self.num % i == 0:
                return False
          
            return True

obj1 = PrimeChecker(10)
result = obj1.is_prime()
print(result)
```

## Task 3: Common Factors Finder


This task finds all common factors between two given numbers.  

```python
class CommonFactorsFinder:
    """
    this class finds all common factors between two numbers.
    
    Attribute_1 : the first number
    type attribute_1 : int
    
    Attribute_2 : the second number
    type attribute_2 : int
     
    """
    def __init__(self , num1 : int , num2 : int):
        self.num1 = num1
        self.num2 = num2
        
    def find_com_facts(self):
        """
        this function finds all common factors between the two numbers.

        return : list of common nums between the two numbers.
        type return : list

        """
        minimum = min(self.num1 , self.num2)
        common = []
    
        for i in range(1, minimum + 1):
            if self.num1 % i == 0 and self.num2 % i == 0:
                common.append(i)
        return common
    
    
obj2 = CommonFactorsFinder(20,4)
res2 = obj2.find_com_facts()
print(res2)
```

## Task 4: Password Generator

This task generates a random password of a fixed length.  

```python
import random

class PasswordGenerator:
    """
   This class generate passord of fixed length.
   
   Attribute_1 : the length of password i want to generate.
   type attribute_1 : int
   
    """
    def __init__(self , length : int):
        self.length = length
        

    def generate_password(self):
        """
        This function is used to generate a random password of given length.
        Includes uppercase, lowercase, digits, and symbols.
        
        return : password generated
        type return : str
        
        """
        chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
        password = ""
    
        for _ in range(self.length):
            password += random.choice(chars)  
        return password


obj3 = PasswordGenerator(8)
res3 = obj3.generate_password()
res3
```

## Task 5: Word Frequency Counter

This task counts how many times each unique word appears in a given list of words.


```python
class WordFrequencyCounter:
    """
    this class counts how many times each unique word appears in a list.
    
    Attribute_1 : list of words i want to calculate wors frequency from.
    type attribute_1 : list
    
    """
    def __init__(self , words_list : list):
        self.words_list = words_list
        
    def count_words(self):
        """
        This function is used for calculating number of every unique word in the list
    
        param_1 : list of words i want to calculate from it 
        type param_1 : list
    
        return : return every unique word with count of it in the list
        type return : dict
    
        """
        unique_words = set(self.words_list)
        count = {word: 0 for word in unique_words}
        for i in unique_words:
            for j in self.words_list:
                if i == j:
                    count[i] += 1
        return count

obj4 = WordFrequencyCounter(["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"])
res4 = obj4.count_words()
res4
```

## Task 6 : Simple ChatBot

This task implements a simple rule-based chatbot.  

```python
import random

class SimpleChatBot:
    """
    A simple chatbot that responds to user input based on predefined responses.
    
    
    """
    def __init__(self):
        
        self.responses = {
            "hello": ["Hello!", "Hi there!", "Greetings!"],
            "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
            "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
            "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }
    
    def get_response(self, user_input):
        """
        Returns a response based on user input.
        
        """
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        return random.choice(self.responses["default"])
    
    def start_chat(self):
        
        print("Hello! I'm a simple ChatBot. Type 'goodbye' to quit.")
        
        while True:
            user_input = input("User: ").lower()
            response = self.get_response(user_input)
            print("ChatBot:", response)
            
            if user_input == "goodbye":
                break

if __name__ == "__main__":
    obj5 = SimpleChatBot()
    obj5.start_chat()
```

## Task 7. 
- Given two strings, write a program that efficiently finds the longest common subsequence

- **Ex:** str1 = "Welcome to Machine Learning Diploma" str2 = "I am studying a Machine Learning Course" --> Machine Learning

```python
class LongestSubsequence:
    """ 
    This class used to find the longest common subsequence between two strings .
    
    """
    def __init__(self, str1: str, str2: str):
        self.str1 = str1
        self.str2 = str2
        
    def long_common( self):
        """ 
        This function is used to find the longest common subsequence between two strings .
        
        param_1 : the first string
        type param_1 : str
        
        param_2 : the second string
        type param_2: str
        
        return : the longest common subsequence between two strings .
        type return : str
        """
    
        longest_str = ""

        for i in range(len(self.str1)):
            for j in range(i, len(self.str1)):
                sub_str = self.str1[i:j+1]
                if sub_str in self.str2:
                    if len(sub_str) > len(longest_str):
                        longest_str = sub_str

        return longest_str
    

obj6 = LongestSubsequence("abcdef", "zcdemf")
print(obj6.long_common())  
```

## Task 8.
- Write a function that will take a given string and reverse the order of words.

- **Ex:** “Hello world” → ”world Hello”


```python
class Reverse_Str:
    """ 
    This class used to reverse a given string.
    
    Attribute_1 : string i want to reverse.
    type attribute_1 : str
    
    """
    def __init__(self , sentence : str):
        self.sentence = sentence
        
    def reverse_sen(self):
        """
        this function take a given string and reverse the order of words.
        
        return : reversed words of string.
        type : str
        
        """
        new_str = self.sentence.split()[::-1]
        new_str = " ".join(new_str)
        return new_str

obj7 = Reverse_Str("Hello World")
obj7.reverse_sen()        
```

## Task 9.
- Write a Python program to find all the unique words and count the frequency of occurrence from a text file and save the counts in a new file
- This is the same as Task 7 but this time read words from a file
- Hint: use replace function and split function

```python
class FileWordFrequencyCounter:
    """
    FileWordFrequencyCounter reads words from a file,
    counts the frequency of each unique word,
    and saves the results into a new file.
    
    Attribute_1 : file that has words
    type attribute_1 : str
    
    Attribute_2 : file i will save the result into
    type attribute_2 : str
    
    """

    def __init__(self, input_file: str, output_file: str):
        self.input_file = input_file
        self.output_file = output_file
    
    def create_sample_file(self):
        """
        Creates a sample input file with some text.
        
        """
        sample_text = "Hello world! Hello Python. Python is great, and Python is fun."
        with open(self.input_file, "w") as file:
            file.write(sample_text)
            
    def count_words(self) -> dict:
        """
        Reads the input file and counts frequency of each unique word.
        
        """
        with open(self.input_file, "r") as file:
            text = file.read().lower()

        # Remove punctuation
        for ch in ",.!?":
            text = text.replace(ch, "")
        words = text.split()

        word_count = {}
        for word in words:
            word_count[word] = word_count.get(word, 0) + 1

        return word_count

    def save_to_file(self):
        """
        Saves the word counts to the output file.
        """
        word_count = self.count_words()
        with open(self.output_file, "w") as file:
            for word, count in word_count.items():
                file.write(f"{word}: {count}\n")


# Example usage
counter = FileWordFrequencyCounter("sample_input.txt", "word_counts.txt")
counter.create_sample_file()      # create a file with sample text
counter.save_to_file()            # count words and save to file
```
