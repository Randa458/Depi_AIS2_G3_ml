# Problem Description – Task 1 (Python Calculator Validation)

## Task Description

- Modify the provided calculator code to validate user inputs using **try-except** blocks.  
- Handle the following cases:  
  1️⃣ Non-numeric input → print a clear error message.  
  2️⃣ Division by zero → print a clear error message.  
  3️⃣ Any unexpected errors → print a general message with the error description.

## Notes
- Make sure function names are correct (e.g., `multiply` instead of `multply`).  
- The goal is to make the program safer and user-friendly, without crashing on errors.


```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    num_1 = float(input("Enter first numb"))
    num_2 = float(input("Enter second numb"))
    
    if choice == "1":
        result = sum(num_1, num_2)
    elif choice == "2":
        result = sub(num_1, num_2)
        
    elif choice == "3":
        result = div(num_1, num_2)
        
    elif choice == "4":
        result = multply(num_1, num_2)
        
    else:
        print("invalid choice")
        return
    
    print("Result: ", result)
    
    
if __name__ == "__main__":
    main()
```

```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multiply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    try:
        choice = input("Enter the operation")
        num_1 = float(input("Enter first numb"))
        num_2 = float(input("Enter second numb"))
    
        if choice == "1":
            result = sum(num_1, num_2)
            
        elif choice == "2":
            result = sub(num_1, num_2)
        
        elif choice == "3":
            result = div(num_1, num_2)
        
        elif choice == "4":
            result = multiply(num_1, num_2)
            
        else:
            print("invalid choice")
            return
        print("Result: ", result)
    
    except ValueError:
        print("Error: Please enter a valid number!")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed!")
    except Exception as e:
        print(f"Unexpected error: {e}")
        
        


if __name__ == "__main__":
    main()
```
