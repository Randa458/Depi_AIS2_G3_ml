```python
import os 
import random

def thanos_project(folder_path : str , num_files : int):
    """
    This function creates a number of files inside a folder, deletes 
    half of them randomly, and prints the remaining files.
    
    param_1 : folder path where files will be created and then deleted randomly.
    type param_1 : str
    
    param_2 : number of files to create inside the folder
    type param_2 : int
    
    return : list of remaining files after random deletion
    type return : list of str
    
    """
    
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")
    else:
        print(f"Folder '{folder_path}' exists.")
        
    for i in range(num_files):
        inner_path = os.path.join(folder_path , f"file_{i+1}.txt")
        with open(inner_path, 'w') as f:
            f.write("") 
        
    files = os.listdir(folder_path)
    print(f"Number of files before deletion: {len(files)}")
    
    files_to_delete = random.sample(files, len(files)//2)
    for file_name in files_to_delete:
        os.remove(os.path.join(folder_path, file_name))
    print(f"Deleted {len(files_to_delete)} random files.")
    
    remaining_files = os.listdir(folder_path)
    print(f"Number of files after deletion: {len(remaining_files)}")
    print("Remaining files:", remaining_files)
    
    return remaining_files
```

```python
thanos_project("ThanosFolder", 10)
```
