```python
def my_filter(fun, lst):
    new_lst =[]
    for i in lst:
        if fun(i):
            new_lst.append(i)
    return new_lst

my_list = [1, 2 ,3 ,4, 5]
my_filter(check_even, my_list)
```

# Scopes

**Global**

```python
x = "global"

def fun(x):
    print("inside:", x)
    
fun(3)
print("outside:", x)
```

**Local**

```python
x = "global"

def fun():
    x = "local"
    print("inside:", x)
    
fun()
print("outside:", x)
```

```python
x = "global"

def fun():
    global x
    x = "changed"
    print("inside:", x)
    
fun()
print("outside:", x)
```

**non-local**

```python
x = "global"

def outer_fun():
    x = "outer_local"
    
    def inner_fun():
        x = "inner_local"
        print("inside inner func:", x)
    
    inner_fun()
    print("inside outer func:", x)
    
outer_fun()
print("outside any func:", x)
```

```python
x = "global"

def outer_fun():
    x = "outer_local"
    
    def inner_fun():
        global x
        x = "changed"
        print("inside inner func:", x)
    
    inner_fun()
    print("inside outer func:", x)
    
outer_fun()
print("outside any func:", x)
```

```python
x = "global"

def outer_fun():
    x = "outer_local"
    
    def inner_fun():
        nonlocal x
        x = "changed"
        print("inside inner func:", x)
    
    inner_fun()
    print("inside outer func:", x)
    
outer_fun()
print("outside any func:", x)
```

# File Handling
### file modes
1. r --> **read mode**: reads the content of the file.
2. a --> **append mode**: creates the file if not exists, without overwriting its content
3. w --> **write mode**: creates the file if not exists, and overwrite its content.
4. x --> **excute mode**: creates the file only, and if it exists it will raise an error.

```python
# Generator: A type of iterable like lists or tuples 
# but does not store the full sequence in memory at once.
# Uses yield to generate one item at a time
#counter generator
def counter(start=0):
    n = start
    while True:
       yield n
       n += 1

for i in counter(5):
    if i > 10:
        break 
    print(i)
```

```python
#Infinite Fibonacci sequence generator
def fib():
    a, b = 0, 1
    while True:
       yield a 
       a, b = b, a + b
       
for n in fib():
   if n > 10:
     break  
   print(n)
```

```python
# Generator expression: More compact syntax like
# list comprehensions for inline lazy generation, uses () instead of []
#Generator expression to get squares
nums = (x**2 for x in range(10))
print(list(nums))

```

```python
# Infinite sequence: Generators can be infinitely recursive/iterative to model data streams
# Infinite random attack sequence
import random
attacks = ["kimura", "armbar", "triangle"] 

def lazy_random_attacks():
    """Lazily yield random attacks forever"""
    
    while True:
        attack = random.choice(attacks)
        print("Yielding attack") 
        yield attack
        
generator = lazy_random_attacks()

for _ in range(5):
    print(next(generator))
```

```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    num_1 = float(input("Enter first numb"))
    num_2 = float(input("Enter second numb"))
    
    if choice == "1":
        result = sum(num_1, num_2)
    elif choice == "2":
        result = sub(num_1, num_2)
        
    elif choice == "3":
        result = div(num_1, num_2)
        
    elif choice == "4":
        result = multply(num_1, num_2)
    else:
        print("invalid choice")
        return
    print("Result: ", result)


if __name__ == "__main__":
    main()
    
```

```python
def reverse_str(s):
        
    lst = s.split(" ")
    new_lst = lst[::-1]
    new_str = " ".join(new_lst)
    return new_str
```

```python
reverse_str("hello world")
```
