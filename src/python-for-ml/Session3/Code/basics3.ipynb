```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b
```

```python
def calculator(x,y, opr ):
  """
  This function works as a caculator to add , divide , substract , multiply 2 numbers from user
  
  param_1 : The first number from user
  type param_1 : int
  
  param_2 : The second number from user
  type param_2 : int 
  
  param_3 : THe operation user want
  type param_3 : str
  
  return : return the result of operation he want
  type return : float
  
  """  
  if opr == "+":
    return add(x, y)
  elif opr == "-":
    return subtract(x, y)
  elif opr == "*":
    return multiply(x, y)
  elif opr == "/":
    return divide(x, y)
  else:
    return "Invalid operation"
  
```

```python
x = int(input("Enter the first num: "))
y = int(input("Enter the second num : "))
opr = input("Enter the operation you want")
result = calculator(x,y,opr)
result
```

```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    num_1 = float(input("Enter first numb"))
    num_2 = float(input("Enter second numb"))
    
    if choice == "1":
        result = sum(num_1, num_2)
    elif choice == "2":
        result = sub(num_1, num_2)
        
    elif choice == "3":
        result = div(num_1, num_2)
        
    elif choice == "4":
        result = multply(num_1, num_2)
    else:
        print("invalid choice")
        return
    print("Result: ", result)


if __name__ == "__main__":
    main()
```

```python
def reverse_words(s : str):
    """
    this function is used to reverse string 
    param_1 : string i want to reverse
    type param_1:str
    return : string reversed
    type return : str
    """
    s = s.split()[::-1]
    new_str = " ".join(s)
    return new_str
```

```python
s = " hello world"
result = reverse_words(s)
result
```
